```python
import tkinter as tk
from tkinter import filedialog, Toplevel
from enum import Enum
import os
from PIL import Image, ImageTk
import random
import matplotlib.pyplot as plt
import cv2
import time

# Define the speed limits for each sign index (excluding 6)
speed_limits = [20, 30, 40, 50, 60, 70, 80, 100, 120]

# Global variables to store weather information
weather_description = ""
temperature = 0.0
weather_image = ""

# Simulate weather data
def simulate_weather():
    weather_conditions = ['Sunny', 'Rainy', 'Cloudy', 'Snowy', 'Windy', 'Foggy']
    weather_images = {
        'Sunny': 'sunny.png',
        'Rainy': 'rainy.png',
        'Cloudy': 'cloudy.png',
        'Snowy': 'snowy.png',
        'Windy': 'windy.png',
        'Foggy': 'foggy.png'
    }
    temperature = random.uniform(-10, 35)  # Temperature range from -10 to 35 degrees Celsius
    weather_description = random.choice(weather_conditions)
    weather_image = weather_images[weather_description]
    return weather_description, temperature, weather_image

def update_weather():
    global weather_description, temperature, weather_image
    weather_description, temperature, weather_image = simulate_weather()

def display_results_in_new_window(image_path, predicted_class_index):
    global weather_description, temperature, weather_image

    # Clear the existing content in the main window
    for widget in root.winfo_children():
        widget.destroy()

    # Load the meta image without the background
    meta_dir = 'C:\\Users\\Lenovo\\Downloads\\archive\\Meta'
    meta_image_path = os.path.join(meta_dir, f"{predicted_class_index}.png")
    if os.path.exists(meta_image_path):
        meta_image = Image.open(meta_image_path)

        # Convert the image to RGBA mode if it's not already in that mode
        if meta_image.mode != 'RGBA':
            meta_image = meta_image.convert('RGBA')

        # Create a new image with a black background (0, 0, 0, 255)
        black_bg = Image.new('RGBA', meta_image.size, (0, 0, 0, 255))
        meta_image_with_black_bg = Image.alpha_composite(black_bg, meta_image)

        # Resize the meta image
        new_size = (55, 55)  # Adjust the size as needed
        meta_image_with_black_bg = meta_image_with_black_bg.resize(new_size)

        # Convert the image back to RGB mode
        meta_image_with_black_bg = meta_image_with_black_bg.convert('RGB')
        meta_image_tk = ImageTk.PhotoImage(meta_image_with_black_bg)

        # Load background image
        background_image = Image.open("C:\\Users\\Lenovo\\Downloads\\Dashboard2.png")
        background_photo = ImageTk.PhotoImage(background_image)
        background_label = tk.Label(root, image=background_photo)

        background_width, background_height = background_image.size
        root.geometry(f"{background_width}x{background_height}")

        background_label.image = background_photo
        background_label.place(x=0, y=0, relwidth=1, relheight=1)

        # Determine position for meta image
        if predicted_class_index <= 8:
            meta_label = tk.Label(root, image=meta_image_tk)
            meta_label.image = meta_image_tk
            meta_label.place(x=358, y=385)  # Coordinates for speed limit signs
        else:
            meta_label = tk.Label(root, image=meta_image_tk)
            meta_label.image = meta_image_tk
            meta_label.place(x=1110, y=385)  # Coordinates for regular signs

        # Display weather information
        weather_label = tk.Label(root, text=f"Weather: {weather_description}, Temp: {temperature:.1f}°C", font=("Arial", 18), bg="black", fg="white")
        weather_label.place(relx=0.43, rely=0.73, anchor="center")

        # Load and display weather image
        weather_img = Image.open(weather_image)
        weather_img = weather_img.resize((40, 40))  # Resize to be small
        weather_photo = ImageTk.PhotoImage(weather_img)
        weather_image_label = tk.Label(root, image=weather_photo)
        weather_image_label.image = weather_photo
        weather_image_label.place(relx=0.56, rely=0.73, anchor="center")

        # Display clock
        clock_label = tk.Label(root, text="", font=("Arial", 18), bg="black", fg="white")
        clock_label.place(relx=0.63, rely=0.73, anchor="center")
        update_clock(clock_label)

        # Display the speed
        vehicle.speed_display = tk.Label(root, text=f"{vehicle.speed}", font=("Arial", 90), bg="black", fg="white")
        vehicle.speed_display.place(relx=0.495, rely=0.48, anchor="center")
        vehicle.update_speed_display()

        # Adjust speed limit
        if predicted_class_index <= 8:
            vehicle.set_limit(speed_limits[predicted_class_index])
        else:
            vehicle.set_limit(None)

    else:
        print(f"Meta image for class {predicted_class_index} not found")

def play_video():
    video_window = Toplevel(root)
    video_window.title("Reverse Camera")

    # Load video file
    video_path = 'C:\\Users\\Lenovo\\Downloads\\archive\\reverse_video.mp4'  # Replace with your video file path
    cap = cv2.VideoCapture(video_path)

    def update_frame():
        ret, frame = cap.read()
        if ret:
            cv2image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGBA)
            img = Image.fromarray(cv2image)
            imgtk = ImageTk.PhotoImage(image=img)
            video_label.imgtk = imgtk
            video_label.configure(image=imgtk)
            video_label.after(10, update_frame)
        else:
            cap.release()
            video_window.destroy()

    video_label = tk.Label(video_window)
    video_label.pack()
    update_frame()

def button_click():
    # Open a file dialog window to choose an image file
    file_path = filedialog.askopenfilename(filetypes=[("Image files", "*.jpg;*.png;*.jpeg")])
    if file_path:
        print("Selected file:", file_path)
        plot, prediction = test_on_img(file_path)
        plt.imshow(plot)
        plt.show()
        predicted_class_index = prediction.argmax()
        predicted_class = classes[predicted_class_index]
        print("Predicted traffic sign is:", predicted_class)
        print(predicted_class_index)
        meta_dir= 'C:\\Users\\Lenovo\\Downloads\\archive\\Meta'
        meta_image= os.path.join(meta_dir, f"{predicted_class_index}.png")
        if os.path.exists(meta_image):
            img = plt.imread(meta_image)
            plt.imshow(img)
            plt.show()
        else:
            print(f"Image for class {predicted_class_index} not found")
        display_results_in_new_window(meta_image, predicted_class_index)

def update_sign():
    # Define the directory where images are stored
    image_dir = 'C:\\Users\\Lenovo\\Downloads\\archive\\Test'
    # Get a list of image files in the directory
    image_files = [os.path.join(image_dir, file) for file in os.listdir(image_dir) if file.endswith(('.jpg', '.png', '.jpeg'))]
    # Choose a random image file from the list
    random_image = random.choice(image_files)
    if random_image:
        plot, prediction = test_on_img(random_image)
        predicted_class_index = prediction.argmax()
        display_results_in_new_window(random_image, predicted_class_index)
    # Schedule the function to be called again after a random interval
    root.after(random.randint(3000, 10000), update_sign)

class GearSelection(Enum):
    """An Enum representing different gear selections in a vehicle."""
    PARK = "P"
    REVERSE = "R"
    NEUTRAL = "N"
    DRIVE = "D"

    @classmethod
    def from_string(cls, value):
        """Converts a string value to the corresponding GearSelection enum member.

        Args:
            value: The string value to convert.

        Returns:
            The GearSelection enum member corresponding to the value, 
            or None if no matching member is found.
        """
        for member in cls:
            if member.value == value:
                return member
        return None

class Vehicle:
    def __init__(self):
        self.current_gear = GearSelection.PARK
        self.gear_display = None
        self.speed = 20  # Initial speed
        self.speed_display = None  # Placeholder for the speed display label
        self.decelerating = False
        self.limit = None

    def set_gear(self, gear):
        if not isinstance(gear, GearSelection):
            raise ValueError("Invalid gear selection. Use GearSelection enum members.")
        self.current_gear = gear
        if self.gear_display:
            self.gear_display.config(text=f"Current Gear: {self.current_gear.value}")

        if self.current_gear == GearSelection.REVERSE:
            play_video()

        if self.current_gear == GearSelection.DRIVE:
            update_sign()
            self.update_speed()

    def set_deceleration(self, decelerate):
        self.decelerating = decelerate

    def set_limit(self, limit):
        self.limit = limit

    def update_speed(self):
        change = 0
        if self.decelerating:
            change = -14  # Always decelerate by 14 km/h
            self.speed = max(0, self.speed + change)
            if self.speed == 0:
                self.decelerating = False  # Stop decelerating when speed is 0
        elif self.limit and self.speed > self.limit:
            change = -14  # Decelerate by 14 km/h if over the limit
            self.speed = max(0, self.speed + change)
        else:
            if self.speed < 80:
                # More likely to accelerate by 14 km/h
                change = random.choices([-14, -3, 0, 3, 14], weights=[1, 1, 1, 1, 4])[0]
            else:
                # More likely to decelerate by 14 km/h
                change = random.choices([-14, -3, 0, 3, 14], weights=[4, 1, 1, 1, 1])[0]
            self.speed = max(0, min(self.speed + change, 200))

        # Update the speed display
        self.update_speed_display()

        # Schedule the next update
        root.after(1000, self.update_speed)

    def update_speed_display(self):
        if self.speed_display:
            if self.limit and self.speed > self.limit:
                self.speed_display.config(text=f"{self.speed}", font=("Arial", 80), bg="black", fg="red")
            else:
                self.speed_display.config(text=f"{self.speed}", font=("Arial", 80), bg="black", fg="white")

def update_clock(clock_label):
    current_time = time.strftime('%H:%M:%S')
    clock_label.config(text=current_time)
    root.after(1000, update_clock, clock_label)

def button_real_time_click(vehicle):
    # Background image
    background_image = Image.open(r'C:\Users\Lenovo\Desktop\depositphotos_113631814-stock-photo-road-or-traffic-signs.jpg')
    background_photo = ImageTk.PhotoImage(background_image)
    background_label.config(image=background_photo)
    background_label.image = background_photo
    background_label.place(x=0, y=0, relwidth=1, relheight=1)
    
    button_park.place(relx=0.1, rely=0.4, anchor="center")
    button_reverse.place(relx=0.1, rely=0.5, anchor="center")
    button_neutral.place(relx=0.1, rely=0.6, anchor="center")
    button_drive.place(relx=0.1, rely=0.7, anchor="center")
    
    vehicle.gear_display.place(relx=0.1, rely=0.3, anchor="center")

    if vehicle.current_gear == GearSelection.DRIVE:
        update_sign()

def return_to_gear_selection():
    # Clear the existing content in the main window
    for widget in root.winfo_children():
        widget.destroy()
    
    # Recreate the gear selection buttons and their placements
    button_park.place(relx=0.2, rely=0.4, anchor="center")
    button_reverse.place(relx=0.2, rely=0.5, anchor="center")
    button_neutral.place(relx=0.2, rely=0.6, anchor="center")
    button_drive.place(relx=0.2, rely=0.7, anchor="center")

    vehicle.gear_display.place(relx=0.5, rely=0.3, anchor="center")


vehicle = Vehicle()

# Create the main application window
root = tk.Tk()
root.title("Traffic Sign Recognition System")

# Set the window size
root.geometry("800x600")

# Background image
background_image = Image.open(r'C:\Users\Lenovo\Desktop\depositphotos_113631814-stock-photo-road-or-traffic-signs.jpg')
background_photo = ImageTk.PhotoImage(background_image)
background_label = tk.Label(root, image=background_photo)
background_label.image = background_photo
background_label.place(x=0, y=0, relwidth=1, relheight=1)

# Initialize the weather widgets
weather_label = tk.Label(root, text="Weather: Fetching...", font=("Arial", 12), bg="black", fg="white")
weather_image_label = tk.Label(root, bg="white")

# Create buttons for gear selection
button_park = tk.Button(root, text="Park (P)", command=lambda: vehicle.set_gear(GearSelection.PARK))
button_reverse = tk.Button(root, text="Reverse (R)", command=lambda: vehicle.set_gear(GearSelection.REVERSE))
button_neutral = tk.Button(root, text="Neutral (N)", command=lambda: vehicle.set_gear(GearSelection.NEUTRAL))
button_drive = tk.Button(root, text="Drive (D)", command=lambda: vehicle.set_gear(GearSelection.DRIVE))

# Create buttons for uploading an image and real-time mode
button_upload = tk.Button(root, text="Upload Image", command=button_click)
button_upload.place(relx=0.3, rely=0.5, anchor="center")

button_real_time = tk.Button(root, text="Real Time Mode", command=lambda: button_real_time_click(vehicle))
button_real_time.place(relx=0.7, rely=0.5, anchor="center")

# Create vehicle gear display
vehicle.gear_display = tk.Label(root, text=f"Current Gear: {vehicle.current_gear.value}", font=("Arial", 12))

# Start fetching weather data once at the beginning
update_weather()

root.mainloop()
```
